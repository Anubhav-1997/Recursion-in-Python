```
def insertion_sort(arr):
  for i in range(1,len(arr)):
    #----------------------------------------------------#
    print("Iteration ",i, "\n") # printing array
    for _i in range(0,len(arr)): # for visualization of swapping 
      print(arr[_i])
    print("\n")
    #----------------------------------------------------#
    key=arr[i][2] # sorting on cities column i.e. 1 
    j=i-1
    keyrow=arr[i]

    while (j >= 0) and (arr[j][2] > key):
      arr[j+1]=arr[j]
      j=j-1

    arr[j+1]=keyrow
#----------------------------------------------------#

lst = data
insertion_sort(lst)
print("Final Sorted Data is below: \n", data)
```
